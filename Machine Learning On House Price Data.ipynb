```python
# In N_Hospital_Beds, we have less than 10% of data is missing
df_House_Price1.dropna(inplace=True)
```

```python
df_target=df_House_Price1['price']
df_feature=df_House_Price1.drop('price',axis=1)
```

```python
# Segregate the data based on the datatype
df_num=df_House_Price1.select_dtypes(include='number')
df_cat=df_House_Price1.select_dtypes(include='object')
```

```python
# Show the numerical data
df_num.head()
```

```python
# Show the categorical data
df_cat.head()
```

### Perform Encoding on categorical variable

```python
# Use Dummy encoding 
df_encoded=pd.get_dummies(df_cat)
# rest the index 
df_encoded=df_encoded.reset_index()
```

```python
# show the encoded data
df_encoded.head()
```

### Perform Scaling on Numerical data 

```python
# import the standard scaler 
from sklearn.preprocessing import StandardScaler

# Initiate the Standard Scaler
ss=StandardScaler()

# Scaled the value present in Numerical dataframe
scaled=ss.fit_transform(df_num)

# Numerate the data into the dataframe 
scaled_df=pd.DataFrame(scaled,columns=df_num.columns)

# Reset the index
scaled_df=scaled_df.reset_index()
```

```python
scaled
```

```python
# show the data
scaled_df.head()
```

### Perform the concatenate operation


```python
# concate encoded variabiles and scaled variables
df_independent=pd.concat([df_encoded,scaled_df],axis=1)
```

```python
# show the data 
df_independent.head()
```

## Split the data for Training and Testing

```python
# import the train test split
from sklearn.model_selection import train_test_split
```

```python
# split the data 
x_train,x_test,y_train,y_test=train_test_split(df_independent,df_target,test_size=0.2)
```

```python
# check the size 
x_train.shape
```

# Apply Linear Regression 

```python
# import the Model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# initiate the model
LR=LinearRegression()
```

```python
# train the liner regression model
model_lr=LR.fit(x_train,y_train)
```

```python
# perform the prediction
predicted=model_lr.predict(x_test)
```
